```python
#https://www.rottentomatoes.com/top/bestofrt/?year=2019
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
```

```python
url = "https://www.rottentomatoes.com/top/bestofrt/?year=2019"
page = urlopen(url)
soup = BeautifulSoup(page, "html.parser")
```

```python
soup.select(".table td")
```

```python
rank = soup.select(".table .bold") 
rating = soup.select(".table .tMeterScore")
title =soup.select(".table .unstyled")
review = soup.select(".table .right")
count = 0

li_count = []
li_Rank = []
li_Rating = [] 
li_Title = []
li_Review = [] 

for i in range(len(rank)):
    count += 1
    li_count.append(count)
    li_Rank.append(rank[i].string.strip())
    li_Rating.append(rating[i].string.strip())
    li_Title.append(title[i].string.strip())
    li_Review.append(review[i+1].string.strip())
```

```python
print(li_Rank)
print(li_Rating)
print(li_Title)
print(li_Review)
      
```

```python
#countLists 만으로 DataFramea 생성
df = pd.DataFrame()
df['Rank'] = pd.Series(li_Rank)
df['Rating'] = pd.Series(li_Rating)
df['Title'] = pd.Series(li_Title)
df['Review'] = pd.Series(li_Review)
df
```

```python
#csvf로 저장
df.to_csv("./Data/RottenTomatoesLists.csv", sep=",", encoding='utf-8', index=False)
```
