# 튜플(Tuple) 자료형
: Tuple은 다음과 같은 점을 제외하고 리스트와 거의 비슷하다.

- (,)로 사용
- 리스트는 그 값을 생성, 삭제, 수정이 가능하지만 튜플은 값을 바꿀수없다.    

```python
# 튜플의 선언
var01 = ()
var02 = (1)
var03 = (1,2,3)
var04 = 1,2,3
var05 = ('a','b',('ab','cd'))
print(var01)
print(var02)
print(var03)
print(var04)
print(var05)
```

```python
print(var03[0])
```

```python
a,b = (1,2)
print(a)
print(b)
```
