# 집합(Set)

```python
# 집합 정의
s1 = set([1,2,3,4,5])
s2 = set([4,5,6,7,8,9])
print(s1)
print(s2)
```

```python
# 교집합
print(s1 & s2)
print(s1.intersection(s2))
```

```python
# 합집합
print(s1 | s2)
print(s1.union(s2))
```

```python
#차집합
print(s1 - s2)
print(s1.difference(s2))
```

```python
# 값 1개 추가
s1.add(10)
s1
```

```python
# 값 여러개 추가
s1.update([20,30,40])
s1
```

```python
#특정값 제거
s1.remove(10)
s1
```
