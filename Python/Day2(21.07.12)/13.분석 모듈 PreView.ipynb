# Dictionary로 DataFrame 만들기

```python
tbl = pd.DataFrame({
    "weight" : [80.0, 70.4, 65.5, 45.9, 51.2],
    "height" : [170, 180, 155, 143, 154],
    "type" : ['f','n','n','t','t']
})
tbl
```

```python
#몸무게 컬럼만 출력하기
tbl.loc[:,'weight']
```

```python
# 몸무게 컬럼과 키 컬럼 출력하기
tbl.loc[:,['weight','height']]
```

```python
# 정렬의 오름차순과 내림차순
# height 기준으로 오름차순
tbl.sort_values(by='height')
```

```python
# height 기준으로 내림차순
tbl.sort_values(by='height', ascending = False)
```

# Pandas와 Matplotlib을 이용하여 시각화 하기

```python
import matplotlib as plt
import pandas as pd
%matplotlib inline
```

```python
# 데이터 불러오기
names = pd.read_csv("./Data/Baby_Names_1880-2014.csv")
```

```python
#데이터를 앞에서 5개만 불러오기
names.head()
```

```python
#데이터를 뒤에서 5개만 불러오기
names.tail()
```

```python
#전체 갯수
names.count()
```

```python
# pivot을 이용하여 각 이름의 count를 년도별 합계로 그룹화 하여 집계하기
total_births = names.pivot_table('births', index='year', columns='gender', aggfunc=sum)
total_births.head(10)
```

```python
#matplotlib을 이용하여 그래프 생성하기
total_births.plot()
```

```python
plt.rcParams['figure.figsize'] = (14, 6)
```

```python
total_births.plot(title = "Total births by gender and year")
```
